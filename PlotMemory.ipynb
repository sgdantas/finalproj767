```python
import numpy as np
import pickle as pk
import matplotlib.pyplot as plt
```

```python
with open('k1memory.pk' , 'rb') as f:
    results = np.array(pk.load(f, encoding='latin1'))
```

```python
with open('k3memory.pk' , 'rb') as f:
    results2 = np.array(pk.load(f, encoding='latin1'))
```

```python
with open('k5memory.pk' , 'rb') as f:
    results3 = np.array(pk.load(f, encoding='latin1'))
```

```python
plt.plot(results,label = 'K=1')
plt.plot(results2,label = 'K=3')
plt.plot(results3,label = 'K=5')
plt.ylabel('Tour length')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
