```python
import seaborn as sns
import numpy as np
import pickle as pk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names = ['k1memory.pk', 'k3memory.pk', 'k5memory.pk']
columns = ['K=1','K = 3', 'K = 5']
df_50 = pd.DataFrame()
df_100 = pd.DataFrame()

for k,name in enumerate(names):
    with open(name , 'rb') as f:
        results_50 = pk.load(f, encoding='latin1')
        df_50[columns[k]] = np.array(results_50)
        
names = ['k1memory_100.pk', 'k3memory_100.pk', 'k5memory_100.pk']

for k,name in enumerate(names):
    with open(name , 'rb') as f:
        results_100 = pk.load(f, encoding='latin1')
        df_100[columns[k]] = np.array(results_100)

df_50 = pd.melt(df_50)
df_50.columns = ['Memory','Average Tour Length']
                 
df_100 = pd.melt(df_100)
df_100.columns = ['Memory','Average Tour Length']
    
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="Memory", y="Average Tour Length",data=df_50).set_title('50 Cities')
plt.show()
```

```python
sns.boxplot(x="Memory", y="Average Tour Length",data=df_100).set_title('100 Cities')
plt.show()
```

```python
names = ['25.pk', '35.pk', '50.pk']
df = pd.DataFrame()

for k,name in enumerate(names):
    with open(name , 'rb') as f:
        results = pk.load(f, encoding='latin1')
        df[name[:-3]+' cities'] = np.array(results)


df = pd.melt(df)
df.columns = ['Number of cities during training', 'Average Tour Length']
```

```python
sns.boxplot(x='Number of cities during training', y='Average Tour Length',data = df).set_title('Tests on 100 cities setting')
```
